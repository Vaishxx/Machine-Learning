```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits #importing dataset from sklearn lib which has predifined dataset of 1797 handwritten digits
```

```python
digits=load_digits() #for loading digits
```

```python
digits.target[:5]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
```

```python
reg.fit(X_train,Y_train)
```

```python
reg.score(X_test,Y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
reg.predict(digits.data[0:5])
```

```python
#confusion matrix : tells us where there's incorrection in the score,why is it so
Y_predict= reg.predict(X_test)
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(Y_test,Y_predict)
cm
```

```python
#for better visualization of the confusion matrix:
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel("predict")
plt.ylabel('Truth')
```

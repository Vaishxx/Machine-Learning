```python
import pandas as pd
```

```python
df=pd.read_csv('titanic_decision_tree_excercise.csv')
df.head(10)
```

```python
#Survived=target
#inputs=sex,age,fare,Pclass
input=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived'],axis=1)
input
```

```python
target=df['Survived']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_pclass=LabelEncoder()
le_sex=LabelEncoder()
```

```python
input['pclass_n']=le_pclass.fit_transform(input['Pclass'])
input['sex_n']=le_pclass.fit_transform(input['Sex'])
input
```

```python
input_n=input.drop(['Pclass','Sex'],axis=1)
input_n
```

```python
input_n['Age'].isnull().sum()
```

```python
mode=input_n['Age'].mode()
mode
```

```python
input_n.Age=input_n['Age'].fillna(mode)
input_n
```

```python
input_n['Age'].isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(input_n,target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```
